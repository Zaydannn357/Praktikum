Dictionary

```python
# a. Buat dictionary keranjang dengan data: ("apel": 5000, "pisang": 3000, "Jeruk": 7000, "anggur": 15000)
keranjang = {
    "apel": 5000,
    "pisang": 3000,
    "Jeruk": 7000,
    "anggur": 15000
}
print(keranjang)

total = 0

# b. Gunakan perulangan untuk menghitung total harga semua barang
for buah, harga in keranjang.items():
    print(f"Harga {buah}: {harga}")
    total += harga
print(f"harga total: {total}")
```
