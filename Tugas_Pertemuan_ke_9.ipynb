```python
# Nama   : M. Gustav Zaydan Nawfal
# NPM    : 2505060060
# Rombel : 03

# 1. Diberikan list berikut:
print('no 1')
buah = ['apel', 'mangga', 'jeruk', 'anggur', 'pisang']
# a. tambahkan 'semangka' di akhir list
buah.append('semangka')
print(f'a. buah setelah ditambah\t: {buah}')
# b sisipkan 'durian' di antara 'jeruk' dan 'anggur'
buah.insert(3, 'durian')
print(f'b. buah setelah disisipkan\t: {buah}')
# c. hapus 'mangga' dari list
buah.remove('mangga')
print(f'c. buah setelah dihapus\t\t: {buah}')
# d. ubah 'pisang' menjadi 'nanas'
buah[4] = 'nanas'
print(f'd. buah setelah diubah\t\t: {buah}')
# e. tampilkan 3 buah pertama
print(f'e. 3 buah pertama\t\t: {buah[:3]}\n')

# 2. Diberikan list berikut:
print('no 2')
angka = [45, 12, 78, 23, 56, 89, 34]
# a. urutkan list secara ascending
angka.sort()
print(f'a. list ascending\t: {angka}')
# b. urutkan list secara descending
print(f"b. list descending\t: {angka}")
```
