```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

# Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

# Function to add newnode in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode

# Function to remove last node
    def RemoveEnd(self):
        last = self.headval

        while (last is not None):
            if last.nextval == None:
                break
            prev = last
            last = last.nextval

        if (last == None):
            return

        prev.nextval = last.nextval
        last = None


# create linked list named Li
Li = LinkedList()

# add node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
print("Sebelum dilakukan remove")
Li.listprint()
print("=============================================")

# remove last node
Li.RemoveEnd()
print("Setelah dilakukan remove node terakhir")
Li.listprint()
```

Double Linked List

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan isi linked list
    def listprint(self):
        print("Isi Linked List:")
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval
        print()

# Fungsi untuk menambah node di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.prevval = last

# Fungsi untuk menyisipkan node setelah kata kunci
    def InsertAfter(self, key, newdata):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                NewNode = Node(newdata)

                NewNode.nextval = temp.nextval
                NewNode.prevval = temp

                if temp.nextval:
                    temp.nextval.prevval = NewNode

                temp.nextval = NewNode
                return

            temp = temp.nextval

# Fungsi untuk menghapus node berdasarkan kata kunci
    def Delete(self, key):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:

                if temp.prevval is None:
                    self.headval = temp.nextval
                    if temp.nextval:
                        temp.nextval.prevval = None
                else:
                    temp.prevval.nextval = temp.nextval
                    if temp.nextval:
                        temp.nextval.prevval = temp.prevval
                return

            temp = temp.nextval


# Membuat linked list bernama Li
Li = DoubleLinkedList()

# Menambah 3 node awal
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
Li.listprint()

print("=============================================")

# Menyisipkan node setelah kata kunci
Li.InsertAfter("Februari", "April")
Li.listprint()

print("=============================================")

# Menghapus node berdasarkan kata kunci
Li.Delete("Februari")
Li.listprint()
```

Circular Linked List

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan isi linked list
    def listprint(self):
        print("Isi Linked List:")
        if self.headval is None:
            return

        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break
        print()

# Fungsi untuk menambah node di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.nextval = self.headval

# Fungsi untuk menghapus node terakhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        if self.headval.nextval == self.headval:
            self.headval = None
            return

        last = self.headval
        while last.nextval.nextval != self.headval:
            last = last.nextval

        last.nextval = self.headval


# Membuat linked list bernama Li
Li = CircularLinkedList()

# Menambah node awal
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
Li.listprint()

print("=============================================")

# Menyisipkan node di akhir
Li.AddInEnd("April")
Li.listprint()

print("=============================================")

# Menghapus node terakhir
Li.RemoveEnd()
Li.listprint()
```
