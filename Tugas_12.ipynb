```python
import math

# a. Fungsi untuk menghitung luas persegi panjang
def luas_persegi_panjang(panjang, lebar):
    return panjang * lebar
p = float(input("Masukkan panjang: "))
l = float(input("Masukkan lebar: "))
print(f"Luas persegi panjang = {luas_persegi_panjang(p, l):.2f}")

# b. Fungsi untuk menghitung volume balok (persegi panjang 3D)
def volume_balok(panjang, lebar, tinggi):
    return panjang * lebar * tinggi
p = float(input("Masukkan panjang: "))
l = float(input("Masukkan lebar: "))
t = float(input("Masukkan tinggi: "))
print(f"Volume balok = {volume_persegi_panjang(p, l, t):.2f}")

# c. Fungsi untuk menghitung luas lingkaran
def luas_lingkaran(jari_jari):
    return math.pi * jari_jari**2
r = float(input("Masukkan jari-jari: "))
print(f"Luas lingkaran = {luas_lingkaran(r):.2f}")

# d. Fungsi untuk menghitung volume bola
def volume_bola(jari_jari):
    return (4/3) * math.pi * jari_jari**3
r = float(input("Masukkan jari-jari: "))
print(f"Volume bola = {volume_bola(r):.2f}")
```
